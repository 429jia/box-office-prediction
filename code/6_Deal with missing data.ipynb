# Clean missing data with mean value

```python
import pandas as pd
import numpy as np
import math
```

```python
movie = pd.read_csv("box2016_mtc_trend_tie copy.csv",index_col=0)
```

```python
movie
```

```python
movie.describe().transpose()
```

```python
x = movie.isnull().sum()
type(x)
```

```python
y = x.to_frame().reset_index()
y.to_csv('missing_value.csv')
```

```python
for each in movie.columns:
    if movie[each].isnull().sum() > 0:
        movie[each] = movie[each].fillna(int(movie[each].mean()))
```

```python
movie.to_csv("box2016_mtc_trend_tie_miss.csv")
```

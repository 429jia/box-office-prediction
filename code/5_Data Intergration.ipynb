# Collect Player Gross

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import csv
```

```python
player = pd.read_csv('people_list.csv', index_col=0)
```

```python
links = player['links']
```

```python
def get_movie(movie_list, link):
    output_list = []
    for movie in movie_list:
        rank = movie.contents[0].text
        if rank.isdigit():
            name = movie.contents[2].text
            name_link = 'http://www.boxofficemojo.com/' + movie.contents[2].font.a['href']
            adjust_gross = movie.contents[6].text.replace(',','').replace('$','')
            unadjust_gross = movie.contents[8].text.replace(',','').replace('$','')
            open_date = movie.contents[10].text
            output_list.append([link, rank, name, name_link, adjust_gross, unadjust_gross, open_date])
    return output_list
```

```python
output_list = []
output_list.append(['link', 'rank', 'name', 'name_link', 'adjust_gross', 'unadjust_gross', 'open_date'])
with open('player.csv', 'a+', encoding='UTF-8', newline='') as csvfile:
    w = csv.writer(csvfile)
    w.writerows(output_list) 
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US; rv:1.9.1.6) Gecko/20091201 Firefox/3.5.6'} 

for i in range(254, len(links)):
    each = links[i]
    output_list = []
    link = "http://www.boxofficemojo.com" + each
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    soup = soup.find('div', id="body").contents[8]

    movie_list1 = soup.find_all('tr', attrs={"bgcolor": "#ffffff"})
    movie_list2 = soup.find_all('tr', attrs={"bgcolor": "#f4f4ff"})
    output_list = output_list + get_movie(movie_list1, link)
    output_list = output_list + get_movie(movie_list2, link)
    with open('player.csv', 'a+', encoding='UTF-8', newline='') as csvfile:
        w = csv.writer(csvfile)
        w.writerows(output_list)
    time.sleep(1)
    print (link)
```

# Merge gross to box data

```python
movie = pd.read_csv('box2016.csv')
```

```python
movie['studio'] = movie['studio'].replace(['BV', 'WB', 'Uni.', 'Fox', 'LG/S', 'Sony', 'STX', 'Par.', 'WB (NL)',
       'Focus', 'LGF', 'SGem', 'TriS', 'RAtt.', 'ORF', 'A24', 'EC', 'BG',
       'QF', 'BH Tilt', 'SPC', 'Free', 'Wein.', 'BST', 'PFR', 'VPD',
       'Rela.', 'FoxS', 'Fathom'], 
    ['Buena Vista', 'Warner Bros.', 'Universal', '20th Century Fox','Lionsgate', 'Sony / Columbia', 'STX Entertainment',
    'Paramount', 'Warner Bros.', 'Focus Features', 'Lionsgate', 'Sony / Columbia', 'Sony / Columbia', 'Roadside Attractions',
    'Open Road Films', 'A24', 'EuropaCorp',
    'Broad Green Pictures', 'Quality Flix', 'BH Tilt', 'Sony Classics',
     'Freestyle Releasing', 'Weinstein Company', 'Bleecker Street',
     'Pure Flix', 'Viva Pictures Distribution', 'Relativity',
     'Fox Searchlight', 'Fathom'
    ])
```

```python
movie['open_date'] =  '2016/'+ movie['open_date']
movie['open_date'] = pd.to_datetime(movie['open_date'])
```

```python
movie['Director_links'] = movie['Director_links'].str.split('|', 1).str[0].str.strip()
movie['Director_list'] = movie['Director_list'].str.split('|', 1).str[0].str.strip()
movie['Actor_links'] = movie['Actor_links'].str.split('|', 1).str[0].str.strip()
movie['Actor_list'] = movie['Actor_list'].str.split('|', 1).str[0].str.strip()
movie.drop(['name_link', 'total_thea','open_gross', 'close_date', 'Writer_list', 'Writer_links','Producer_list', 'Producer_links', 'Cinematographer_list',
       'Cinematographer_links', 'Composer_list', 'Composer_links'], axis=1, inplace=True)
movie['MPAArating'] = movie['MPAArating'].replace('R',1).replace('PG-13',0).replace('PG',0)
movie['budget'] = movie['budget'].str.replace('$','').str.replace('million','')
movie['runtime'] = movie['runtime'].str.replace(' min.','')
movie['runtime'] = movie['runtime'].str.split('hrs.', 1).str[0].astype(int)*60 + movie['runtime'].str.split('hrs.', 1).str[1].astype(int)

movie['Sci-Fi'] = movie['genre'].str.contains('Sci-Fi')
movie['Adventure'] = movie['genre'].str.contains('Adventure')
movie['Action'] = movie['genre'].str.contains('Action')
movie['Animation'] = movie['genre'].str.contains('Animation')
movie['Musical'] = movie['genre'].str.contains('Musical')
movie['Comedy'] = movie['genre'].str.contains('Comedy')
movie['Fantasy'] = movie['genre'].str.contains('Fantasy')
movie['Thriller'] = movie['genre'].str.contains('Thriller')
movie['Horror'] = movie['genre'].str.contains('Horror')
movie['Drama'] = movie['genre'].str.contains('Drama')
movie['Romance'] = movie['genre'].str.contains('Roman')
movie['Historical'] = movie['genre'].str.contains('Historical')
movie['Western'] = movie['genre'].str.contains('Western')
movie['Family'] = movie['genre'].str.contains('Family')
movie['Crime'] = movie['genre'].str.contains('Crime')
movie['Concert'] = movie['genre'].str.contains('Concert')
movie['Documentary'] = movie['genre'].str.contains('Documentary')
movie['Sports'] = movie['genre'].str.contains('Sports')
movie['War'] = movie['genre'].str.contains('War')
movie['Documentary'] = movie['genre'].str.contains('Documentary')
movie[['Sci-Fi', 'Adventure','Action', 'Animation', 'Musical', 'Comedy', 'Fantasy', 'Thriller',
       'Horror', 'Drama', 'Romance', 'Historical', 'Western', 'Family','Crime', 'Concert', 'Documentary', 'Sports', 'War']] =movie[['Sci-Fi', 'Adventure','Action', 'Animation', 'Musical', 'Comedy', 'Fantasy', 'Thriller',
       'Horror', 'Drama', 'Romance', 'Historical', 'Western', 'Family','Crime', 'Concert', 'Documentary', 'Sports', 'War']].astype(int)
```

```python
movie_player = pd.merge(movie, clean_player_sum,  how='left', left_on=['Director_links'], right_on = ['link'])
movie_player2 = pd.merge(movie_player, clean_player_sum,  how='left', left_on=['Actor_links'], right_on = ['link'])
movie_player2.rename(columns={'adjust_gross_x': 'Director_gross', 'adjust_gross_y': 'Actor_gross'}, inplace=True)
movie_player2.drop(['genre','Director_links','Actor_links','link_x','link_y'], axis=1, inplace=True)
```

```python
movie_player2
```

```python
studio = pd.read_csv('studio.csv')
studio['total_gross'] = studio['total_gross'].str.replace(',','')
studio['total_gross'] = pd.to_numeric(studio['total_gross'])
studio = studio[studio['year'].isin(['2015','2014','2013','2012','2011'])]
studio2 = studio.groupby('distributor').sum().reset_index()
studio2.drop(['year', 'rank','movies_tracked','movies_num'], axis=1, inplace=True)
studio2.columns = ['studio2', 'studio_gross']
movie_player_studio = pd.merge(movie_player2, studio2,  how='left', left_on=['studio'], right_on = ['studio2'])
movie_player_studio.drop(['studio','studio2'], axis=1, inplace=True)
```

```python
movie_player_studio.to_csv('box2016_clean.csv')
```

# Merge Metascore with box

```python
mc = pd.read_csv("mc2_2016.csv", index_col=0)
```

```python
mc["moviename"] = mc["moviename"].fillna('')
mc["moviename_x"] = mc["moviename_x"].fillna('')
mc["box_moviename"] = mc["moviename"].map(str) + mc["moviename_x"]
mc['date'] = pd.to_datetime(mc['date'])
mc.drop(['moviename','moviename_x','author_link','mixed','negative','num_crit','positive','positive','source_link','summary','release_date'], axis=1, inplace=True)
```

```python
mc2 = pd.merge(mc, movie_player_studio,  how='left', left_on=['rank'], right_on = ['rank'])
```

```python
mc2 = mc2[['date','iscore','meta_movie_name','rank','name','open_date','box_moviename']]
```

```python
mc2['date_diff'] = mc2['open_date'] - mc2['date']
mc2['date_diff'] = mc2['date_diff'] / np.timedelta64(1, 'D')
```

```python
mc2
```

```python
mc_before = mc2[mc2['date_diff']>0]
```

```python
grouped = mc_before.groupby(['rank'])
mc_avg = grouped['iscore'].agg(['mean', 'count']).reset_index()
```

```python
mc_avg = mc_avg.rename(columns={'mean': 'mean_score_mtc', 'count': 'comment_count_mtc'})
```

```python
movie_player_studio_mc = pd.merge(movie_player_studio, mc_avg,  how='left', left_on=['rank'], right_on = ['rank'])
```

# Merge google trend with box

```python
google_trend = pd.read_csv("google_trend.csv",index_col=0)
movie_player_studio_mc = pd.read_csv("box2016_clean_mtc.csv",index_col=0)
google_name = pd.read_csv("google_name.csv")
google_name.drop([ 'name'], axis=1, inplace=True)
```

```python
movie_mc = pd.merge(movie_player_studio_mc, google_name,  how='left', left_on=['rank'], right_on = ['rank'])
```

```python
movie_mc
```

```python
movie_mc['trend_a_week'] = pd.Series()
movie_mc['trend_two_week'] = pd.Series()
movie_mc['trend_a_day'] = pd.Series()
```

```python
for i in range(0,len(movie_mc)):
    if type(movie_mc.iloc[i]['ID']) != float:
        index = google_trend['date'][google_trend['date'] == movie_mc.iloc[i]['open_date']].index[0]
        aweek = google_trend.iloc[index-7:index][movie_mc.iloc[i]['ID']].mean()
        twoweek = google_trend.iloc[index-14:index-7][movie_mc.iloc[i]['ID']].mean()
        aday = google_trend.iloc[index-1][movie_mc.iloc[i]['ID']]
        movie_mc['trend_a_week'][i] = aweek
        movie_mc['trend_two_week'][i] = twoweek
        movie_mc['trend_a_day'][i] = aday
```

```python
movie_mc
```

```python
movie_mc.to_csv('box_2016_trend.csv')
```
